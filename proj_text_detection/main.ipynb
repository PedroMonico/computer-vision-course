```python
## Text Detection using Tesseract ##

import pytesseract
from PIL import Image

image_path = './data/i_see_a_light_in_the_darkness.jpg'

text = pytesseract.image_to_string(Image.open(image_path), lang='eng')
print(text)
```

```python
## Text Detection using EasyOCR ##

from easyocr import Reader

reader = Reader(['en'])

results = reader.readtext(Image.open(image_path))
#print(results)

text = ''
for result in results:
    text = text + result[1] + ' '

text = text[:-1]  # removes the white space at the end
print(text)
```

```python
## Compare Perfomance
import os
from jaccrd_sim import jaccard_similarity
from easyocr import Reader

image_path_ = './data'

def read_text_tesseract(image_path):
    text = pytesseract.image_to_string(Image.open(image_path), lang='eng')
    return text

def read_text_easyocr(image_path):
    reader = Reader(['en'])
    results = reader.readtext(Image.open(image_path))

    text = ''
    for result in results:
        text = text + result[1] + ' '
    text = text[:-1]
    return text

score_tesseract = 0
score_easyocr = 0

for image_path_ in os.listdir('./data'):
    image_path = os.path.join('./data', image_path_)

    text = image_path_[:-4].replace('_', ' ').lower()

    score_tesseract += jaccard_similarity(read_text_tesseract(image_path).lower().replace('\n', '').replace('!', '').replace('?', ''), text)
    score_easyocr += jaccard_similarity(read_text_easyocr(image_path).lower().replace('\n', '').replace('!', '').replace('?', ''), text)


print('score tesseract:', score_tesseract / 100)
print('score easyocr:', score_easyocr / 100)
```
